```python
sales_numbers = np.random.randint(low = 0, high = 30, size = 336)
unit = 10000
monthly_sales = sales_numbers.reshape([12,4,7])
weekly_sales = sales_numbers.reshape([48,7])
```

```python
np.sum(weekly_sales)*unit
```

```python
total_daily_sales_for_fiscal_year = np.sum(weekly_sales, axis=0)
```

```python
plt.bar("mon,tue,wed,thur,fri,sat,sun".split(","),total_daily_sales_for_fiscal_year)
```

```python
total_daily_sales_for_fiscal_year = np.sum(weekly_sales, axis=1)
```

```python
enumed_vals = list(enumerate(total_daily_sales_for_fiscal_year))
preprocessed_values = list(map(lambda x: [f"wk{x[0]}",x[1]], enumed_vals))
```

```python
np_weekly_sales = np.array(preprocessed_values)
qtys = np_weekly_sales[:,1]
wks = np_weekly_sales[:,0]
np_weekly_sales
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(20,6),dpi=80,facecolor='w',edgecolor='k')
plt.bar(list(wks),qtys.astype('int'), color='k')
```
