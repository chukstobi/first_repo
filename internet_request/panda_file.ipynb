```python
import requests
api_key = '605a3f6ecadf2aba2db67fdf2721361f'
city_id = '2332453'
response = (f"https://api.openweathermap.org/data/2.5/forecast?id={city_id}&APPID={api_key}")
fetch = requests.get(response)
wea = fetch.json()
```

```python
data_dict ={
    "time":[],
    "temp K":[],
    "temp c":[],
    "windspeed":[],
    "weather":[]
}

for forecast in wea["list"]:
    data_dict["time"].append(forecast["dt_txt"])
    data_dict["temp K"].append(forecast["main"]["temp"])
    data_dict["temp c"].append(forecast["main"]["temp"]-273)
    data_dict["windspeed"].append(forecast["wind"]["speed"])
    data_dict["weather"].append(forecast["weather"][0]["description"])

```

```python
import pandas as pd
```

```python
lagos_weather = pd.DataFrame(data_dict)
```

```python
min_temp = lagos_weather["temp c"].min()
```

```python
lagos_weather[lagos_weather["temp c"] == min_temp]
```
