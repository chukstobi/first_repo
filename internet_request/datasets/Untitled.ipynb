```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
glob = pd.read_csv("global_results.csv")
lag = pd.read_csv("results.csv")
```

```python
lag = lag[["year", "avg_temp"]]
```

```python
lag.dropna(0, inplace = True)
```

```python
lag
```

```python
plt.plot(lag.year, lag.avg_temp)
```

```python
m_avg = []
n = 6

padding = [lag.avg_temp.mean()]*n
new_avg = [*padding,*(list(lag.avg_temp))]

def moving_avg(sample, n):
    if len(sample)<n+1:
        print("done")
        return
    chunk = sum(sample[:n])/n
    m_avg.append(chunk)
    sample.pop(0)
    moving_avg(sample, n)
    
    
moving_avg(new_avg, n)
```

```python
plt.plot(lag.year, m_avg)
plt.plot(lag.year, lag.avg_temp)
```
