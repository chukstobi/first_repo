```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
data = pd.read_csv("IRIS.csv")
data
```

```python
sns.pairplot(data, hue="species")
```

```python
ml_features = data[["sepal_length","sepal_width","petal_length","petal_width"]]
ml_labels = data["species"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(ml_features,ml_labels,test_size=0.1,random_state=1)
```

```python
model = KNeighborsClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
output_data = pd.DataFrame({})
output_data["Actual"] = y_test
output_data["Predicted"] = y_pred
```
